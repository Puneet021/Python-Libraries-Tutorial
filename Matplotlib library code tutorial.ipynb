### which one should you use? (pyplot vs matplotlib OO method?)

* when we use to plot quickly better to use pyplot
* when plotting something more advanced use OOmethod

```python
over_50 = heart_disease[heart_disease["age"] >50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol'
            ,c='target');
```

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax = ax);
#ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
           y=over_50["chol"],
           c=over_50["target"]);
 
# Customize the plot
ax.set(title="Heart disease and cholestrol level",
       xlabel="Age",
       ylabel="cholestrol");

# Add a legend 
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10)
                               ,sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# customize data
ax0.set(title="Heart disease and Cholestrol level", ylabel="cholestrol")

# add a legend
ax0.legend(*scatter.legend_elements(),
          title="Target")

# add a horizontal line
ax0.axhline(over_50["chol"].mean(), ls='--');

# add data to ax1
scatter1 = ax1.scatter(x=over_50["age"]
                      ,y=over_50["thalach"],
                       c=over_50["target"])

#customize the data
ax1.set(title="Heart disease and max heart rate",
         xlabel="Age",
        ylabel="Max Heart Rate")

#add a legend
ax1.legend(*scatter1.legend_elements(), title="Target")

# make a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            ls='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
           y=over_50["chol"],
           c=over_50["target"],
                    cmap="winter");
 
# Customize the plot
ax.set(title="Heart disease and cholestrol level",
       xlabel="Age",
       ylabel="cholestrol");

# Add a legend 
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10)
                               ,sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"],cmap="winter");

# customize data
ax0.set(title="Heart disease and Cholestrol level", ylabel="cholestrol")

# change the x axis limits
ax0.set_xlim([50, 80])
# add a legend
ax0.legend(*scatter.legend_elements(),
          title="Target")

# add a horizontal line
ax0.axhline(over_50["chol"].mean(), ls='--');

# add data to ax1
scatter1 = ax1.scatter(x=over_50["age"]
                      ,y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")

#customize the data
ax1.set(title="Heart disease and max heart rate",
         xlabel="Age",
        ylabel="Max Heart Rate")

ax1.set_ylim([60, 200])
#add a legend
ax1.legend(*scatter1.legend_elements(), title="Target")

# make a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            ls='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
